```python
pop = np.random.normal(10,20,100000)
pop
```

# Central limit theorem

```python
sample = [rd.sample(pop.tolist(),k=30) for i in range(20)]
sample
```

```python
samples = np.array(sample)
samples.shape
```

```python
sample_X = samples.mean(axis = 1)
sample_X
```

```python
u = sample_X.mean()
```

```python
u
```

```python
sample_std = samples.std(axis =1)
sample_std.shape
```

```python
std = np.mean(sample_std)
std
```

```python
# Confidence Interval
ci = [u - (t)*(std/(30**(1/2))), u + (t)*(std/(30**(1/2)))]
ci
```
