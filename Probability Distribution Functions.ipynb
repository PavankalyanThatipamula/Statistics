```python
population = np.random.normal(loc = 10,scale = 2,size = 100000)
```

```python
samples =[]
for x in range(20):
    samples.append(rd.sample(population.tolist(),k= 40))
    
```

```python
samplef = np.array(samples)
samplef.shape
```

```python
# samplef.mean(axis = 1).shape
sample_mean = samplef.mean(axis = 1)
```

```python
sns.kdeplot(x = sample_mean)
plt.axvline(sample_mean.mean(),color = 'r',linestyle = ':')
```

```python
sample_mean.mean()
```

### Normal , Uniform Continuous, Lognormal

```python
x =np.random.normal(loc = 0,scale = 1,size = 1000000)
x
```

```python
sns.kdeplot(x)
```

```python
s.mean(x)
```

```python
s.median(x)
```

```python
ss.skew(x)
```

```python
ss.kurtosis(x)
```

```python
## Q-Q plot
```

```python
y= np.random.normal(loc = 0,scale = 1,size = 1000000)
```

```python
xp = np.percentile(x,range(1,100))
yp = np.percentile(y,range(1,100))
```

```python
# sns.scatterplot
sns.regplot(x = xp,y = yp)
plt.xlabel('Sample Distribution of Normal')
plt.ylabel('Theoritical Distribution of Normal')
plt.title('Q-Q plot')
```

```python
#other way of finding q-q plot(quartile -quartile plot)
```

```python
sm.qqplot(x,dist= ss.norm,fit = True,line = "45")
```

## LogNormal Distribution

```python
x2 = np.random.lognormal(mean = 30,sigma =0.5,size = 1000000)
x2
```

```python
sns.kdeplot(x2)
```

```python
x2_log = np.log(x2)
```

```python
sm.qqplot(x2_log,dist = ss.norm,fit = True,line = "45")
plt.show()
```

```python
sm.qqplot(x2,dist = ss.lognorm,fit =True,line = "45")
plt.show()
```
