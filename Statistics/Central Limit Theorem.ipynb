```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
import scipy as s
import random as rd
import statsmodels.api as sm
```

# Central Limit Theorem

```python
population = np.random.normal(10,1,1000000)
population
```

```python
sns.kdeplot(population)
```

```python
samples  = []
for i in range(10):
    samples.append(rd.sample(population.tolist(),k= 30))
samples
```

```python
samples1 = np.array(samples)
samples1.shape
```

```python
samples1.mean(axis = 1).shape
```

```python
sample_mean = samples1.mean(axis =1)
```

```python
sns.kdeplot(sample_mean)
```

```python
sample_mean.mean()
```
